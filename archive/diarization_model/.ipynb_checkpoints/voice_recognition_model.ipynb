```python
import time
import csv

# List of wav files without extension
wav_files = ['mock_call_19', 'mock_call_15']

auth_token = "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"  # Set up your token in Hugging Face


for audio_file_name in wav_files:

    # audio_file = path # Update with your actual file path
    # audio_file_name = '0638'   #edit with your audio file
    audio_file = path + f'/data/{audio_file_name}.mp3'
    print(audio_file)

    output_csv = path + f"/output/output-{audio_file_name}.csv"  # Specify the CSV file path

    # Start timer
    start_time = time.time()

    # Process the audio and get the results
    result = process_audio(audio_file, auth_token)

    # End timer and calculate the elapsed time
    elapsed_time = time.time() - start_time
    print(f"Processing time for {audio_file_name}: {elapsed_time:.2f} seconds")

    # Check if the result is valid before attempting to print and save
    if result:
        print("Saving the final result to CSV...")
        try:
            with open(output_csv, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerow(["Start Time", "End Time", "Speaker", "Transcript"])
                for seg, spk, sent in result:
                    line = f'{seg.start:.2f} {seg.end:.2f} {spk} {sent}'
                    # print(line)
                    writer.writerow([f'{seg.start:.2f}', f'{seg.end:.2f}', spk, sent])
            print(f"Results successfully saved to {output_csv}")
        except Exception as e:
            print(f"Error saving CSV file: {e}")
    else:
        print("No result to print or save.")
```
