# Load Data

```python
data = pd.read_csv(path)
```

```python
N, n = data.shape
print(data.shape)
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
for c in data['Category'].unique():
    
    l = len(np.where((data['Person'] == 'A') & (data['Category'] == c))[0])
    #l = len(np.where(data['Category'] == c)[0])
    
    print('{}: {}'.format(c, l))
    
```

```python
data = data[['Category', 'Text']].reset_index(drop=True)
```

```python
#df.fruit = pd.factorize(df.fruit)[0]

data.Category = pd.factorize(data.Category)[0]
```

```python
data
```

```python
labels = []
texts = []
for i in range(0, data.shape[0]):
    texts.append(data.loc[i][1])
    labels.append(data.loc[i][0])
```

```python
texts
```

```python
# Make sure length of reviews is only 200 words
maxlen = 20

# We will select this number of training data to learn, quite few, you can change if you want
training_samples = 800 
validation_samples = 200

# Number of total words in the dictionary has to be 10000, 
# Considers only the top 10,000 words in the dataset
max_words = 10000 

# Number of embedding dimensions
embedding_dims = 300
```

```python
tokenizer = keras.preprocessing.text.Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
```

```python
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
word_index
```

```python
data = keras.preprocessing.sequence.pad_sequences(sequences, maxlen=maxlen)
```

```python
labels = np.asarray(labels)
print('Shape of data tensor:', data.shape)
print('Shape of label tensor:', labels.shape)
```

```python
# Randomize
indices = np.arange(data.shape[0])
np.random.shuffle(indices)
data = data[indices]
labels = labels[indices]

# Split into training and validation
x_train = data[:training_samples]
y_train = labels[:training_samples]
x_val = data[training_samples: training_samples + validation_samples] 
y_val = labels[training_samples: training_samples + validation_samples]
```

```python
y_val.shape
```

```python
y_train
```

```python
y_train_binary = to_categorical(y_train)
y_val_binary = to_categorical(y_val)
```

```python
y_train_binary
```

```python
y_val_binary
```

```python
x_train
```

***

```python
model = keras.models.Sequential() 

model.add(keras.layers.Embedding(max_words, 32)) 
model.add(keras.layers.SimpleRNN(32)) 

model.add(keras.layers.Dense(7, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])

history = model.fit(x_train, y_train_binary,
                    epochs=20,
                    batch_size=32,
                    validation_data=(x_val, y_val_binary))

model.save_weights('models/embeddings_model.h5')
```

```python
embedding_dim = 200
```

```python
embeddings_index = {}
f = open(os.path.join(glove_dir, 'glove.6B.{}d.txt'.format(embedding_dim)))
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
print('Found %s word vectors.' % len(embeddings_index))
```

```python
embedding_matrix = np.zeros((max_words, embedding_dim))
for word, i in word_index.items():
    if i < max_words:
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
```

```python
model = keras.models.Sequential() 

model.add(keras.layers.Embedding(max_words, embedding_dim)) 
model.add(keras.layers.SimpleRNN(32)) 

model.add(keras.layers.Dense(7, activation='softmax'))
model.summary()
```

```python
model.layers[0].set_weights([embedding_matrix])
model.layers[0].trainable = False
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])

history = model.fit(x_train, y_train_binary,
                    epochs=50,
                    batch_size=32,
                    validation_data=(x_val, y_val_binary))

model.save_weights('models/embeddings_model.h5')
```
