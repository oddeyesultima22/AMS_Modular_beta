```python
# Make sure length of reviews is only 200 words
maxlen = 20

# We will select this number of training data to learn, quite few, you can change if you want
training_samples = 800 
validation_samples = 200

# Number of total words in the dictionary has to be 10000, 
# Considers only the top 10,000 words in the dataset
max_words = 10000 

# Number of embedding dimensions
embedding_dims = 300
```

# 2.1 Model Building (Embeddings)

```python
NUM_EXP = 10

TRAIN_LOSS = []
VAL_LOSS = []
TEST_LOSS = []
TRAIN_ACC = []
VAL_ACC = []
TEST_ACC = []

CM = {}

for exp in range (0, NUM_EXP):
    
    print(' -------------------------------------------------------------------------------------------------------------')
    print('Round {}'.format(exp))
    print(' -------------------------------------------------------------------------------------------------------------')
        
    # Randomize
    indices = np.arange(data.shape[0])
    np.random.shuffle(indices)
    data = data[indices]
    labels = labels[indices]

    # Split into training and validation
    x_train = data[:training_samples]
    y_train = labels[:training_samples]
    x_val = data[training_samples: training_samples + validation_samples] 
    y_val = labels[training_samples: training_samples + validation_samples]
    x_test = data[training_samples + validation_samples:] 
    y_test = labels[training_samples + validation_samples:]

    y_train_binary = to_categorical(y_train)
    y_val_binary = to_categorical(y_val)
    y_test_binary = to_categorical(y_test)

    print('Train set size: {}'.format(x_train.shape))
    print('Validation set size: {}'.format(x_val.shape))
    print('Test set size: {}'.format(x_test.shape))
    
    # Lets add an embedding Layer
    model = keras.models.Sequential()

    model.add(keras.layers.Embedding(max_words, embedding_dims, input_length=maxlen))

    # Lets add a Dense Layer
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(32, activation='relu'))
    model.add(keras.layers.Dense(6, activation='softmax'))
    #model.summary()
    
    early_stopper = keras.callbacks.EarlyStopping(monitor="val_acc", min_delta=0.1, patience=10)
    checkpoint = keras.callbacks.ModelCheckpoint(filepath_checkpoints, monitor="val_acc", verbose=0, save_best_only=True, mode="max")

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])

    history = model.fit(x_train, y_train_binary,
                        epochs=20,
                        batch_size=32,
                        validation_data=(x_val, y_val_binary),
                        #callbacks=[callback], restore_best_weights=True)
                        callbacks = [early_stopper, checkpoint, tfdocs.modeling.EpochDots()])

    hist = pd.DataFrame(history.history)
    hist["epoch"] = history.epoch

    # save training history
    hist.to_csv(os.path.join(LogsDir, "history.csv"))

    model.reset_metrics()
    model.save_weights(os.path.join(ModelDir, "emulator.h5"))

    model.load_weights(filepath_checkpoints)
    
    train_loss, train_acc = model.evaluate(x_train, y_train_binary)
    val_loss, val_acc = model.evaluate(x_val, y_val_binary)
    test_loss, test_acc = model.evaluate(x_test, y_test_binary)
    
    TRAIN_LOSS.append(train_loss)
    TRAIN_ACC.append(train_acc)
    VAL_LOSS.append(val_loss)
    VAL_ACC.append(val_acc)
    TEST_LOSS.append(test_loss)
    TEST_ACC.append(test_acc)
    
    prediction_test = model.predict(x_test)
    prediction_test_results = np.argmax(prediction_test, axis=1)
    
    cm = confusion_matrix(y_test, prediction_test_results, [0, 1, 2, 3, 4, 5])
    
    CM[exp] = cm
    
print('\n')
print('Train: Mean = {}, STD = {}'.format(np.mean(TRAIN_ACC), np.std(TRAIN_ACC)))
print('Validation: Mean = {}, STD = {}'.format(np.mean(VAL_ACC), np.std(VAL_ACC)))
print('Test: Mean = {}, STD = {}'.format(np.mean(TEST_ACC), np.std(TEST_ACC)))
print('\n')

temp_cm = CM[0]

for i in range(1, NUM_EXP):
    temp_cm += CM[i]
    
final_cm = np.ceil(temp_cm/NUM_EXP)
    
```

```python
labels = ["Greetings", "Engagement", "Problem Investigation", "Problem Resolution", "Closure", "Account Verification"]

#sns.heatmap(cm, cmap='Blues')
#make_confusion_matrix(cm, categories=labels, cmap='binary')
make_confusion_matrix(final_cm, categories=labels, figsize=(18,9))
```

```python
embedding_dim = 200
```

```python
embeddings_index = {}
f = open(os.path.join(glove_dir, 'glove.6B.{}d.txt'.format(embedding_dim)))
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
print('Found %s word vectors.' % len(embeddings_index))
```

```python
embedding_matrix = np.zeros((max_words, embedding_dim))
for word, i in word_index.items():
    if i < max_words:
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Embedding(max_words, embedding_dim, input_length=maxlen))

model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(32, activation='relu'))
model.add(keras.layers.Dense(7, activation='softmax'))
model.summary()
```

```python
model.layers[0].set_weights([embedding_matrix])
model.layers[0].trainable = False
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])

history = model.fit(x_train, y_train_binary,
                    epochs=50,
                    batch_size=32,
                    validation_data=(x_val, y_val_binary))

model.save_weights('models/embeddings_model.h5')
```
